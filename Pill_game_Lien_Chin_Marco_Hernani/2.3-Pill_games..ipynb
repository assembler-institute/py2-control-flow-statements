## First game: Guess a number!

**Guess the number!**

Hi, User!


This game consists of choosing a number between one to ten (both included). If you get it right, you win.
How many tries will you need? 


Good luck!

```python
from random import randint

def guess_number(n_succes):
    '''Gets a number from the user and tells you if it's the same than number passed as argument'''
    n_entry = int(input ('Enter a number: '))
    if n_entry == n_succes: 
        print('You are right!')
        return True 
    else: 
        print('Please, try again')
        return False


#randint generates a number random int from one to ten
#both included
n_choosen = randint(1,10)

correct = False 
n_tries = 0

#While correct is False, the loop repeats itself
while not correct:
    #Asks the user to make a guess
    #and the correct variable is updated
    #with the result
    correct = guess_number(n_choosen)
    n_tries += 1
    #If the user has tried 5 times a
    #encouraging message is printed
    if n_tries == 5:
        print("¡Don't give up!")
    
print(f'You tried {n_tries} times until guessing')
```






## Second game: Rock paper scissors

Hi, User!

This game is the classic **rock, paper, scissors** game but digital. It’s simple and fun.

**Here is how to play:**

> 1. Make a choice between three options: rock/paper/scissors
> 2. The computer will also make his choice.  
> 3. The program will tell you if you won, lost or if you drew. Remember:
>- Rock beats scissors
>- Scissors beats paper
>- Paper beats rock.

> 4. Repeat 2 more times.The one who wins more times wins the game!

Enjoy it!


```python
from random import choice

#For every choice this dictionary gives you as value the choice that loses against that choice
choice_to_win = {"rock": "scissors",
         "paper": "rock",
         "scissors": "paper"
}

def rps_fight():
    '''The function will asks the user to choice and will return whether the user has won, lost or the game has been tied.'''
    #User makes his choice
    user_choice = input('Introduce an option rock/scissors/paper: ')
    #choice will get randomly an item of the list
    pc_choice = choice(['rock','scissors','paper'])

    print(f"Computer choice: {pc_choice}")
    
    #If they draw returns 0
    #if the user wins returns 1
    #if the user losts returns -1
    if user_choice == pc_choice:
        print("Draw")
        return 0
    #was the play that loses against the user made by the computer?
    #If so the user wins
    elif choice_to_win[user_choice] == pc_choice:
        print('You have won') 
        return 1
    else:
        print("You have lost") 
        return -1

#Stores the sum of all the fights results
result = 0

for i in range(3):
    #A new fight is made and the result is added to variable result
    result += rps_fight()

#If result is positive it means user wins
#If is negative it means he lost
#else they drew
if result > 0: 
    print('You won the game!!!')
elif result < 0:
    print('You lost the game')
else:
    print('Game tied')


    


```

##  Third game: Rock paper scissors lizard Spock 


Hi, User!


This is rock paper scissors lizard Spock game. It's an extended version of the classic game rock paper scissors.

**Instructions** are as follows:

> 1. Make a choice between five options: rock/paper/scissors/lizard/Spock
> 2. The computer will also make his choice. 
> 3. The program will tell you if you won, lost or if you drew. Remember:

<img width=500 src="Images/relations_rpslS_game.jpg">

>- Scissors cuts paper
>- Paper covers rock
>- Rock crushes lizard
>- Lizard poisons Spock
>- Spock smashes scissors
>- Scissors decapitates lizard
>- Lizard eats paper
>- Paper disproves Spock
>- Spock vaporizes rock
>- Rock crushes scissors.

> 4. Repeat 2 more times.The one who wins more times wins the game!




Are you ready?


**Have fun!**


```python
from random import choice

#Dictionary with the relation between a choice and the choices that loses against that choice
choices_to_win = {"rock": ("lizard", "scissors"),
         "paper": ("rock", "Spock"),
         "scissors": ("paper", "lizard"),
         "lizard": ("Spock", "paper"),
         "Spock": ("scissors", "rock")
}

def fight():
    '''The function will asks the user to choice and will return whether the user has won, lost or the game has been tied.'''
    #User makes his choice
    user_choice = input('Introduce an option rock/scissors/paper/lizard/Spock: ')
    #choice will get randomly an item of the list
    pc_choice = choice(['rock','scissors','paper', 'lizard','Spock'])

    print(f"Computer choice: {pc_choice}")
    
    #If they draw returns 0
    #if the user wins returns 1
    #if the user losts returns -1
    if user_choice == pc_choice:
        print("Draw")
        return 0
    #was the play that loses against the user made by the computer?
    #If so, the user wins
    elif pc_choice in choices_to_win[user_choice]:
        print('You have won') 
        return 1
    else:
        print("You have lost") 
        return -1

#Stores the sum of all the fights results
result = 0

for i in range(3):
    #A new fight is made and the result is added to variable result
    result += fight()

#If result is positive it means user wins
#If is negative it means he lost
#else they drew
if result > 0: 
    print('You won the game!!!')
elif result < 0:
    print('You lost the game')
else:
    print('Game tied')


    


```
