## Guess a number
The game consists of choosing a number from 1 to 10, and see if you have hit the same as the computer.
If you get it right, you win!


```python
#Definimos la función
def guess_number(): 
#Importamos random para obtener números aleatorios
    import random  
#Establecemos el rango de valores que podemos introducir    
    n=random.randrange(1,10)
    nu=int(input("Tell me the number you think I've chosen: "))
#Utilizamos el while para que si no ha adivinado, sigamos jugando hasta que acertemos    
    while nu!=n:
        if nu>n:
            nu=int(input("The number is smaller: "))
        elif nu<n:
            nu=int(input("The number is bigger: "))
    print("Congratulations you guessed the secret number: ",n)
#Llamamos a la función    
guess_number()
```

## Rock Paper Scissors
This is the classic game of rock, paper or scissors. 
The rules are:
* Stone breaks the scissors
* Scissors cuts the paper
* Paper wraps the stone

```python
#Creación de una clase para las opciones ya que nos permite empaquetar datos y funcionalidades juntas

import random

#Para obtener la enumeración basada en un número entero
from enum import IntEnum

#Asi creamos un nuevo tipo de objeto (introduciendo class). 
#De ésta manera, en vez de tener que colocar la palabra, realizamos una equivalencia entre la palabra y el número
class Action(IntEnum):
    Rock = 0
    Paper = 1
    Scissors = 2

#Definimos la función para pedirle al usuario un número
def get_user_selection():
    choices = [f"{action.name}[{action.value}]" for action in Action]
    choices_str = ", ".join(choices)
    selection = int(input(f"Enter a choice ({choices_str}): "))
    action = Action(selection)
    return action

#Determinamos otra función utilizando random para que el ordenador nos devuelva x número
def get_computer_selection():
    selection = random.randint(0, len(Action) - 1)
    action = Action(selection)
    return action

#Esta última función determina quien es el ganador
def determine_winner(user_action, computer_action):
    if user_action == computer_action:
        print(f"Both players selected {user_action.name}. It's a tie!")
    elif user_action == Action.Rock:
        if computer_action == Action.Scissors:
            print("Rock smashes scissors! You win!")
        else:
            print("Paper covers rock! You lose.")
    elif user_action == Action.Paper:
        if computer_action == Action.Rock:
            print("Paper covers rock! You win!")
        else:
            print("Scissors cuts paper! You lose.")
    elif user_action == Action.Scissors:
        if computer_action == Action.Paper:
            print("Scissors cuts paper! You win!")
        else:
            print("Rock smashes scissors! You lose.")

#Creamos un bucle para preguntar al usuario si quiere continuar jugando
while True:
    try:
        user_action = get_user_selection()
    except ValueError as e:
        range_str = f"[0, {len(Action) - 1}]"
        print(f"Invalid selection. Enter a value in range {range_str}")
        continue

    computer_action = get_computer_selection()
    determine_winner(user_action, computer_action)

    play_again = input("Play again? (yes/no): ")
    if play_again.lower() != "yes":
        break
```

## Rock paper scissors lizard Spock
It is a variation of the previous game, but where there are less chances of a draw, because spock and lizard come into play.
The rules are:
* Scissors cut the paper
* Paper covers the stone,
* Stone crushes the lizard,
* Lizard poisons Spock,
* Spock crushes the scissors, 
* Scissors decapitate the lizard, 
* Lizard eats the paper,
* Paper refutes Spock,
* Spock vaporizes the stone,
* Stone crushes the scissors.

```python
#Con el molde del caso anterior, lo que hacemmos es sólo agregar dos variables más: lizard and spock
import random
from enum import IntEnum

class Action(IntEnum):
    Rock = 0
    Paper = 1
    Scissors = 2
    Lizard = 3
    Spock = 4
# Creamos un diccionario donde determinamos para cada elemento, 
#quien gana a quien, y dentro del diccionario, los elementos ganadores, son dispuestos en listas.
victories = {
    Action.Scissors: [Action.Lizard, Action.Paper],
    Action.Paper: [Action.Spock, Action.Rock],
    Action.Rock: [Action.Lizard, Action.Scissors],
    Action.Lizard: [Action.Spock, Action.Paper],
    Action.Spock: [Action.Scissors, Action.Rock]
}



def get_user_selection():
    choices = [f"{action.name}[{action.value}]" for action in Action]
    choices_str = ", ".join(choices)
    selection = int(input(f"Enter a choice ({choices_str}): "))
    action = Action(selection)
    return action

def get_computer_selection():
    selection = random.randint(0, len(Action) - 1)
    action = Action(selection)
    return action

def determine_winner(user_action, computer_action):
    defeats = victories[user_action]
    if user_action == computer_action:
        print(f"Both players selected {user_action.name}. It's a tie!")
    elif computer_action in defeats:
        print(f"{user_action.name} beats {computer_action.name}! You win!")
    else:
        print(f"{computer_action.name} beats {user_action.name}! You lose.")

# Introducimos un while para continuar en el juego    
while True:
    try:
        user_action = get_user_selection()
# Añadimos  la excepción de que si el jugador se equivoca y pone un número diferente del 0-4,
#lo tenga en cuenta y le avise que se ha equivocado de número y que eliga otro                
    except ValueError as e:
        range_str = f"[0, {len(Action) - 1}]"
        print(f"Invalid selection. Enter a value in range {range_str}")
        continue

    computer_action = get_computer_selection()
    determine_winner(user_action, computer_action)
    
#Para preguntar si quiere seguir jugando o no
    play_again = input("Play again? (yes/no): ")
    if play_again.lower() != "yes":
        break
```
